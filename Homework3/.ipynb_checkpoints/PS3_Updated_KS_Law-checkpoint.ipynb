# Introduction - the Kennicutt-Schmidt Law

For Paper #1, we are investigating the long-observed correlation between the star formation rate (SFR) surface density and the gas surface density in galaxies. Star formation is a complex physical process, whose intricacies cannot be captured in a simple analytic expression, **however** we can parametrize its global relationship with the ISM density empirically over an entire galaxy using scaling laws. Schmidt (1959) proposed a power-law relationship between the surface density of the SFR ($\Sigma_{\rm{SFR}}$) and the surface density of the gas ($\Sigma_{\rm{gas}}$). In this law, both quantities are integrated measurements over entire galaxies.

$\Sigma_{\rm{SFR}}$ = A ($\Sigma_{\rm{gas}}$)$^\rm{n}$

Kennicutt (1998) found that quiescent spiral galaxies, IR-luminous starbursts, and circumnuclear starbursts obey a tight relationship defined by the above equation, with a power-law index of n = 1.4 $\pm$ 0.15.

In our class 'Journal Club,' we will be discussing the seminal [Kennicutt 1998 paper](https://ui.adsabs.harvard.edu/abs/1998ApJ...498..541K/abstract). However, this is now quite out of date. Many works exist on this topic since (and before!) then. A recent paper by [de los Reyes & Kennicutt (2019)](https://https://ui.adsabs.harvard.edu/abs/2019ApJ...872...16D/abstract) revisits this law with a larger sample and more accurate data. We are going to use data from this paper to study this correlation.

```python
import astropy.units as u
import astropy.constants as c
from astropy.coordinates import SkyCoord, search_around_sky
from astropy.time import Time
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob
#%matplotlib notebook
%matplotlib inline
from astropy.io import fits
from itertools import combinations
import pickle
from astropy.cosmology import WMAP9 as cosmo
import seaborn as sns
from scipy import stats

plt.rcParams['figure.figsize'] = (10, 10)
plt.rc('axes', labelsize=27)
#plt.rc('axes', labelweight='bold')
plt.rc('axes', titlesize=27)
plt.rc('axes', titleweight='bold')
plt.rc('font', family='sans-serif')
plt.rcParams.update({'font.size': 20})
```

Importing the data:

```python
#I was lazy and just downloaded the whole table. I'm displaying the keywords here so I can look back at them.
hdu_list = fits.open('asu(1).fit')
t = Table(hdu_list[1].data)
t.keys()
```

pulling from your data format:

```python
#Adding the .value will return the elements as an array and not the column. This can turn in to a mess later if you
#try to do any algebra with these arrays.

tlogSFR = t['logSigSFRc'].value
tlogHI = t['logSigHI'].value
tlogH2 = t['logSigH2'].value
mtype=t['MType'].value

#I removed the np.where functions because they are reduntant here. You are indexing with a boolean array
#so the where functions are just kind of an extra step you don't need to do. Same thing when you make the 
#scatter plot. You don't need the np.where() function. 

logSFR = tlogSFR[((tlogH2!=0) & (tlogSFR!=0))]
logHI = tlogHI[((tlogH2!=0) & (tlogSFR!=0))]
logH2 = tlogH2[((tlogH2!=0) & (tlogSFR!=0))]
mtype = mtype[((tlogH2!=0) & (tlogSFR!=0))]
```

```python
#Changing this keyword will change the type of objects we plot. It will make reproducing the code a bit easier.
typ = 's'
#Indexing the data, combining HI and HII:
h1h2 = np.log10(10**(logHI[mtype==typ]) + 10**(logH2[mtype==typ]))
sfr = logSFR[mtype==typ]
#The scatter plot
plt.scatter(h1h2, sfr, label = 'Spirals', alpha = 0.75, edgecolor = 'k', c = 'r', zorder = 10)
#Fitting the line

#There seem to be some NaN values we need to get rid of. You can easily invert a boolean array with "~"
#so we will invert a boolean check for Nan values to get rid of them. This wasn't important for the scatter
#plot because matplotlib will default to ignore them.
mask = ~np.isnan(h1h2) & ~np.isnan(sfr)
#Getting the line from 
m, c, r_value, p_value, std_err = stats.linregress(h1h2[mask], sfr[mask])

plt.plot(h1h2[mask], h1h2[mask]*m+c, c = 'r', lw = 5)
plt.text(-.5,-4, f'y = {round(m,2)}x {round(c,2)}', rotation = 46, c = 'r', zorder = 25)

#Stealing your labels
plt.xlabel(r'Log $\Sigma_{\rm{H_1 + H_2}}$ [M$_\odot$ pc$^{-2}$]') 
plt.ylabel(r'Log $\Sigma_{\rm{SFR}}$ [M$_\odot$ year$^{-1}$ kpc$^{-2}$]') 
plt.title('Updated Kennicutt-Schmidt Law \n Spirals')
plt.xlim(-1, 3)
plt.ylim(-4.9,1)

plt.legend(loc='lower right')

plt.grid(zorder = 0)
```
