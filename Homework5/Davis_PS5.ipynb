### Problem 3a (7pts):  
Assume that a gas cloud with a mass of 10^9  M$_\odot$ collapses instantaneously (yes, this
is unphysical, but it's an idealized case, so let's go with it), and stars are formed with masses
distributed according to a Salpeter initial mass function (power law index = 2.35), with
a minimum mass of 0.08  M$_\odot$ and a maximum of 100  M$_\odot$. Calculate the number of stars of
each spectral type, using the following definitions: O=20-100 M$_\odot$; B=3-20  M$_\odot$; A=1.7-3  M$_\odot$;
F=1.1-1.7  M$_\odot$; G=0.8-1.1  M$_\odot$; K=0.6-0.8  M$_\odot$; M=0.08-0.6  M$_\odot$. Write these values down in
tabular form and show all work.

$M_{tot} = 10^9 M_{\odot}$

index = -2.35

We start with the salpeter equation:

$ \xi (m)\Delta m=\xi _{0}\left({\frac {m}{M_{\odot }}}\right)^{-2.35}\left({\frac {\Delta m}{M_{\odot }}}\right)$

We first need to solve for the normalization constant.

$M_{tot} = \int_{0.08 M_{sun}}^{100 M_{sun}} \xi _{0} \frac{m^{-2.35}}{m} dm$


$M_{tot} = \xi_0 \frac{1}{M_\odot}[\frac{m^{-0.35}}{-0.35}]|_{0.08 M_\odot}^{100 M_\odot}$

$M_{tot} = \xi_0 \frac{1}{-0.35}[100 M_\odot^{-0.35} - 0.08 M_\odot^{-0.35}]$

$\xi_0 = \frac{M_{tot} }{ \frac{1}{-0.35}[100 M_\odot^{-0.35} - 0.08 M_\odot^{-0.35}]}$

```python
#3a if you choose to include anything here (not required)
# Table showing number of stars in each spectral type

import astropy.units as u
import astropy.constants as c
import pandas as pd


msun = c.M_sun.value
mtot = 1e9*msun

#Value in kg
xi0 = mtot/( (1/(-0.35))*( ((100*msun)**(-0.35)) - ((0.08*msun)**(-0.35))  ))

print(f'The normalization constant is {xi0:.2e} in terms of solar masses.')
```

```python
def ngal(lower_m, upper_m):
    """
    Takes a range of masses and computes the number of galaxies within the range.
    
    """
    
    return xi0*(1/msun)*( (1/(-0.35))*( ((upper_m*msun)**(-0.35)) - ((lower_m*msun)**(-0.35))  ))
```

```python
no = ngal(20,100)
print(f'There are {no:.2e} O type stars.')



nb = ngal(3,20)
print(f'There are {nb:.2e} B type stars.')



na = ngal(1.7,3)
print(f'There are {na:.2e} A type stars.')



nf = ngal(1.1,1.7)
print(f'There are {nf:.2e} F type stars.')



ng = ngal(0.8,1.1)
print(f'There are {ng:.2e} G type stars.')



nk = ngal(0.6,0.8)
print(f'There are {nk:.2e} K type stars.')



nm = ngal(0.08,0.6)
print(f'There are {nm:.2e} M type stars.')

nstars = pd.DataFrame({'Types': ['O', 'B', 'A', 'F', 'G', 'K', 'M'],
                     'Number': [no,nb,na,nf,ng,nk,nm]})
nstars
```

### Problem 3c (6 pts): 
Using the results from part a, now calculate the integrated spectrum of your entire
stellar population. Assume that all O stars have the spectrum of an O5V, all B stars have
B5V, etc. Plot this.

```python
# Code for 3c

#For each of these, I will integrate numerically using the trapezoid rule in python. 
#I can do this simply with the numpy library.

tots = []

spec_keys = Kmodels.keys()[1:8]
num_keys = nstars['Types']
numbers = nstars['Number']
spec_arrays = []
for i in range(len(num_keys)):
    int_one_star = np.trapz(Kmodels[spec_keys[i]].values, wavelength)
    all_stars = numbers[i] * int_one_star
    tots.append(all_stars)
    spec_arrays.append(numbers[i] * Kmodels[spec_keys[i]].values)
    print(f'For {num_keys[i]} type stars, we measure {all_stars:.2e} total flux [nJy].')
```

```python
plt.rcParams['figure.figsize'] = (15, 10)
int_spec = np.sum(spec_arrays,axis = 0)
plt.plot(wavelength, int_spec)

plt.yscale('log')
plt.title('Integrated Total Spectra \n d = 10 pc')
plt.xlabel('Wavelength [$\AA$]')
plt.ylabel('Flux [nJy]');
```
